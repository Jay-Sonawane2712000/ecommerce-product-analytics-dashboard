# 01 Data Inspection

This notebook checks the structure of the electronics store ecommerce dataset before any cleaning or analysis.

## 1. Import libraries and define the expected dataset path

We load the basic libraries we need and point to the expected CSV file inside the `data` folder.

```python
from pathlib import Path
import pandas as pd

DATA_PATH = Path("..") / "data" / "events.csv"
print(f"Expected dataset path: {DATA_PATH.resolve()}")
```

## 2. Check that the dataset file exists

If the file is missing, this cell will stop with a clear message so we know exactly what to place in the `data` folder.

```python
if not DATA_PATH.exists():
    raise FileNotFoundError(
        "Dataset file not found. Expected file: "
        f"{DATA_PATH.resolve()}\n"
        "Please place the electronics store dataset CSV in the data folder and rerun the notebook."
    )
```

## 3. Load the dataset

We read the CSV into a pandas DataFrame so we can inspect its structure.

```python
df = pd.read_csv(DATA_PATH)
print("Dataset loaded successfully.")
```

## 4. Preview the first few rows

This gives us a quick look at the raw records and column layout.

```python
df.head()
```

## 5. Check the dataset shape

This shows how many rows and columns are in the dataset.

```python
print("Shape:", df.shape)
```

## 6. List the column names

This helps confirm the exact field names we will use later.

```python
print("Columns:")
for col in df.columns:
    print(f"- {col}")
```

## 7. Inspect data types

This helps us see which columns may need conversion later, especially dates and numeric values.

```python
print(df.dtypes)
```

## 8. Check missing values by column

This gives an early view of data completeness.

```python
missing_values = df.isna().sum().sort_values(ascending=False)
print(missing_values)
```

## 9. Check the unique values in `event_type`

This confirms the available event stages in the dataset.

```python
if "event_type" in df.columns:
    print(sorted(df["event_type"].dropna().unique().tolist()))
else:
    print("Column 'event_type' was not found in the dataset.")
```

## 10. Check the date range of `event_time`

We convert `event_time` to datetime and inspect the earliest and latest timestamps.

```python
if "event_time" in df.columns:
    event_time = pd.to_datetime(df["event_time"], errors="coerce")
    print("Minimum event_time:", event_time.min())
    print("Maximum event_time:", event_time.max())
else:
    print("Column 'event_time' was not found in the dataset.")
```
